```python
import pandas as pd
import numpy as np
from selenium import webdriver
```

```python
#next G0iuSb
url = "https://www.google.com/search?q=the+new+york+times+opinion+politics&rlz=1C1CHBD_enIN829IN829&sxsrf=ALeKk01B6ceXU2qWMZlR53QNkv28NeAFGQ:1585829328560&source=lnms&tbm=nws&sa=X&ved=2ahUKEwjRm7bO2snoAhUMyDgGHdctBA8Q_AUoAXoECA4QAw&biw=1366&bih=608"
driver = webdriver.Chrome('C:/Users/Admin/Desktop/chromedriver')
driver.get(url)
html = driver.page_source.encode('utf-8')
page_num = 0
import time
titles_final = []

while driver.find_elements_by_css_selector('#pnnext'):
    driver.find_element_by_css_selector('#pnnext').click()
    page_num += 1
    print("getting page number "+str(page_num))
    time.sleep(10)
    div = driver.find_elements_by_class_name("dO0Ag")
    for a in div:
        titles_final.append(a.text)
        print(a.text)
        print("-"*35)
```

```python
mylist = list(dict.fromkeys(titles_final))
```
